```python
def categorize_age(age):
    if age < 20 or age > 80:
        return "Out of Range"
    elif 20 <= age <= 29:
        return "20-29"
    elif 30 <= age <= 39:
        return "30-39"
    elif 40 <= age <= 49:
        return "40-49"
    elif 50 <= age <= 59:
        return "50-59"
    elif 60 <= age <= 69:
        return "60-69"
    elif 70 <= age <= 80:
        return "70-80"
    else:
        return "Invalid Age"

# Apply the function to the 'Age' column and create a new column 'Age Category'
df['Age Category'] = df['Age'].apply(categorize_age)

# Display the updated DataFrame
print(df)
```

### 2. RISK ASSESSMENT

```python
# Define high-risk policyholders based on claim frequency (>=2) or claim amount (>50,000)
df['high_risk'] = ((df['claim_freq'] >= 2) | (df['claim_amt'] > 50000)).astype(int)

# Group by multiple factors (household_income, education, coverage_zone) and calculate the mean risk
risk_assessment = df.groupby(['household_income', 'education', 'coverage_zone'])['high_risk'].mean().reset_index()

# Print the results to check the analysis
print(risk_assessment.head())
```

```python
# Define high-risk policyholders based on claim frequency (>=2) or claim amount (>50,000)
df['high_risk'] = ((df['claim_freq'] >= 2) | (df['claim_amt'] > 50000)).astype(int)

# Group by factors to calculate the average risk by income, education, and coverage zone
risk_by_income = df.groupby('household_income')['high_risk'].mean().reset_index()
risk_by_education = df.groupby('education')['high_risk'].mean().reset_index()
risk_by_coverage = df.groupby('coverage_zone')['high_risk'].mean().reset_index()

# Print the results to check the analysis
print(risk_by_income.head())
print(risk_by_education)
print(risk_by_coverage)

# Plot the factor importance

# Plot for Household Income vs High Risk
plt.figure(figsize=(10, 6))
sns.scatterplot(x='household_income', y='high_risk', data=risk_by_income)
plt.title('High-Risk Policyholders by Household Income')
plt.xlabel('Household Income')
plt.ylabel('Proportion of High-Risk Policyholders')
plt.show()

# Plot for Education Level vs High Risk
plt.figure(figsize=(8, 6))
sns.barplot(x='education', y='high_risk', data=risk_by_education)
plt.title('High-Risk Policyholders by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Proportion of High-Risk Policyholders')
plt.xticks(rotation=45)
plt.show()

# Plot for Coverage Zone vs High Risk
plt.figure(figsize=(8, 6))
sns.barplot(x='coverage_zone', y='high_risk', data=risk_by_coverage)
plt.title('High-Risk Policyholders by Coverage Zone')
plt.xlabel('Coverage Zone')
plt.ylabel('Proportion of High-Risk Policyholders')
plt.xticks(rotation=45)
plt.show()
```

### Risk Assessment:
### Insight: The factors most indicative of high-risk policyholders include:
### Household Income: Policyholders with lower household incomes tend to be higher risk, with higher claim frequencies.
### Education Level: Those with lower education levels (e.g., high school graduates) show a higher likelihood of being high-risk compared to those with bachelor's or higher degrees.
### Coverage Zone: Policyholders in high-risk zones (e.g., highly urban or rural areas) tend to file more frequent claims and have higher claim amounts.
### Frequent Claimants: Policyholders who make frequent claims share characteristics such as lower income, higher risk coverage zones, and lower education levels.


### CUSTOMER SEGMENTATION AND MARKETING

### CUSTOMER SEGMENT BASE

```python
# Define the top 25% of household income as high-income threshold
income_threshold = df['household_income'].quantile(0.75)  # Adjust threshold as needed

# Filter policyholders with low claim frequency and high household income
low_claim_high_income = df[(df['claim_freq'] == 0) & (df['household_income'] >= income_threshold)]

# Group by education, coverage zone, and marital status to identify key segments
segmentation = low_claim_high_income.groupby(['education', 'coverage_zone', 'marital_status']).size().reset_index(name='count')

# Display segmentation results
print(segmentation.head())
```

```python
# Visualize segmentation by education, coverage zone, and marital status
plt.figure(figsize=(10, 6))
sns.barplot(x='education', y='count', hue='coverage_zone', data=segmentation)
plt.title('Segmented High-Value Customers by Education and Coverage Zone')
plt.xlabel('Education Level')
plt.ylabel('Count of Policyholders')
plt.xticks(rotation=45)
plt.show()
```

### 3. Customer Segmentation and Marketing:
#### Insight: Policyholders with low claim frequency and high household income tend to:
#### Be highly educated, often holding bachelor’s or master’s degrees.
#### Live in urban or suburban coverage zones, where claim frequency is lower.
#### Be predominantly married, suggesting that marital stability could be a factor contributing to lower risk behavior.
#### Marketing Strategy: Focus marketing efforts on attracting and retaining high-value customers from higher income brackets and urban areas, particularly those with higher education and married status, as these groups tend to have lower claim frequencies.

### GEOGRAPHICAL ANALYSIS

```python
# 1. Analyze Claim Frequency and Amount by Coverage Zone

# Group by coverage zone to calculate average claim frequency and claim amount
zone_analysis = df.groupby('coverage_zone').agg(avg_claim_freq=('claim_freq', 'mean'),avg_claim_amt=('claim_amt', 'mean')).reset_index()
print(zone_analysis)

# Plot average claim frequency by coverage zone
plt.figure(figsize=(8, 6))
sns.barplot(x='coverage_zone', y='avg_claim_freq', data=zone_analysis)
plt.title('Average Claim Frequency by Coverage Zone')
plt.xlabel('Coverage Zone')
plt.ylabel('Average Claim Frequency')
plt.xticks(rotation=45)
plt.show()

# Plot average claim amount by coverage zone
plt.figure(figsize=(8, 6))
sns.barplot(x='coverage_zone', y='avg_claim_amt', data=zone_analysis)
plt.title('Average Claim Amount by Coverage Zone')
plt.xlabel('Coverage Zone')
plt.ylabel('Average Claim Amount')
plt.xticks(rotation=45)
plt.show()
```

### 5. Geographical Analysis:
#### Insight: Claim frequencies and amounts vary significantly across coverage zones:
#### Highly Urban Zones: Show the highest claim frequencies, likely due to the dense population and higher risk of accidents or theft.
#### Rural Zones: While rural areas have lower claim frequencies, the claim amounts tend to be higher when claims are made, likely due to factors such as limited access to services or higher repair costs.
#### Suburban and Urban Zones: These zones display a balance with moderate claim frequencies and claim amounts, suggesting they are lower-risk compared to highly urban and rural areas.
#### Regional Trends: For marketing or risk assessment, urban zones represent lower-risk, high-value opportunities, while rural zones need to be monitored for high-cost claims despite lower frequencies.


### CUSTOMER BEHAVIOR

```python
# Create a new column to indicate whether the policyholder has kids driving
df['has_kids_driving'] = df['kids_driving'] > 0

# Group by 'has_kids_driving' to analyze the impact of children driving on claim frequency and amount
kids_driving_analysis = df.groupby('has_kids_driving').agg(avg_claim_freq=('claim_freq', 'mean'),avg_claim_amt=('claim_amt','mean')).reset_index()
print(kids_driving_analysis)

# 1. Plot average claim frequency by whether there are kids driving
plt.figure(figsize=(8, 6))
sns.barplot(x='has_kids_driving', y='avg_claim_freq', data=kids_driving_analysis)
plt.title('Average Claim Frequency: Presence of Kids Driving')
plt.xlabel('Has Kids Driving')
plt.ylabel('Average Claim Frequency')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

# 2. Plot average claim amount by whether there are kids driving
plt.figure(figsize=(8, 6))
sns.barplot(x='has_kids_driving', y='avg_claim_amt', data=kids_driving_analysis)
plt.title('Average Claim Amount: Presence of Kids Driving')
plt.xlabel('Has Kids Driving')
plt.ylabel('Average Claim Amount')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

### 6. Customer Behavior Insights:
#### Insight: The presence of children driving significantly impacts claim frequency and amounts:
#### Higher Claim Frequency: Policyholders with children driving tend to have higher claim frequencies, likely due to the increased risk of younger or less experienced drivers on the road.
#### Claim Amount: The average claim amount for policyholders with children driving is also higher, indicating that accidents involving younger drivers could result in more significant claims.
#### Actionable Insights: Insurers may consider adjusting premiums for policyholders with children driving, or offering driver education programs to help mitigate this increased risk.
